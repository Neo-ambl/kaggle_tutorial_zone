```python
# --- 提出ファイル作成用の別ファイルに書くコード ---
import pandas as pd
import numpy as np

# 1. 予測に必要な「元のtestデータ（idが含まれているもの）」を読み込む
# ※パスはご自身の環境（ローカルにアップロードした場所）に合わせてください
test = pd.read_csv('../../data/raw/test.csv') 

# 2. ステップ1で保存した「予測確率」のファイルを読み込む
pred_proba = np.loadtxt('../../model/pred_test_proba_age_flag.txt')

# 3. 提出用ファイルの作成（指定のフォーマット「id,Exited」に合わせる）
submission = pd.DataFrame({
    'id': test['id'],
    'Exited': pred_proba
})

# 4. ローカル環境にCSVとして書き出し
submission.to_csv('../../results/submission_age_flag.csv', index=False)

print("提出用ファイル 'submission_age_flag.csv' を作成しました！")
# 中身の形式が正しいか先頭5行を表示して確認
print(submission.head())
```
